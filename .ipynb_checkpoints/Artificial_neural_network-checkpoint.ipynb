Our third model is an ANN. We will do a simple neural network with multiclass classification using softmax regression and cross-entropy loss as cost function.


We will use a portion of the famous MNIST dataset.

To compare we will build an ANN in TensorFlow



```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import fetch_openml
```

# Preparing Data

We will load the data from sklearn. It will return a dataframe

```python
digits = fetch_openml('mnist_784')
```

```python
X = digits.data
Y = digits.target
```

```python
X
```

### Visualize one digit

```python
digit = np.array(X.loc[15])
plt.imshow(digit.reshape(28,28),cmap ='binary')
plt.show()
```

## Make a subset of data

For this first approach we will be using 2000 examples from the original dataset to speed the training

```python
y_subset = Y[:2000]
X_subset = X[:2000]
```

```python
y_subset.value_counts().plot(kind = 'bar')
plt.show()
```

Every class is covered. Move on!

1. Convert dataframes to arrays.
2. Create train and test sets

```python
X_arr = np.array(X_subset)
y_arr = np.array(y_subset)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_arr,y_arr,random_state = 0,test_size = 0.2)
```
